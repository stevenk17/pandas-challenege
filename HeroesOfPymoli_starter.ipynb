```python
# Dependencies and Setup
import pandas as pd 
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load) 

purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate number of unqiue items
unique = purchase_data["Item Name"].nunique()
#Calculate average price
avg_price = np.mean(purchase_data["Price"])
#Calculate purchases amount
num_purchases = purchase_data["Item Name"].size
#Calculate Rev. 
revenue = sum(purchase_data["Price"])

#Display table 
table = pd.DataFrame([{"Number of Unique Items": unique, "Average Price": avg_price, "Number of Purchases": num_purchases,
                       "Total Revenue": revenue}])

table["Average Price"] = table["Average Price"].map("${:.2f}".format)

table["Total Revenue"] = table["Total Revenue"].map("${:,.2f}".format)

table = table[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Bin
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100] 

demographics = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

purchase_data["Age Demographic"] = pd.cut(purchase_data["Age"], bins, labels = demographics) 

age_demographics = purchase_data.groupby("Age Demographic") 

#Number of purchases per demographic 
purchase_count = age_demographics["Purchase ID"].count() 

#Price per demographic 
avg_price = age_demographics["Price"].mean() 

#Amount purchased per demographic
total_purchase_value = purchase_count * avg_price 

avg_total_per_person = age_demographics["Price"].sum()/age_demographics["SN"].nunique() 

table_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_price, "Total Purchase Value": total_purchase_value, "Average Total per Person": avg_total_per_person}) 

table_df["Average Purchase Price"] = table_df["Average Purchase Price"].map("${:.2f}".format) 

table_df["Average Total per Person"] = table_df["Average Total per Person"].map("${:.2f}".format) 

table_df["Total Purchase Value"] = table_df["Total Purchase Value"].map("${:.2f}".format) 

table_df = table_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Average Total per Person"]] 

table_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#calculations

spending = purchase_data.groupby("SN")

num_purchases = spending["Purchase ID"].count()

total_purchase_value = spending["Price"].sum()

avg_price = total_purchase_value/num_purchases

#Display table

table_df = pd.DataFrame({"Purchase Count": num_purchases, "Average Purchase Price": avg_price, 
                           "Total Purchase Value": total_purchase_value})
table_df = table_df.sort_values(by="Total Purchase Value", ascending = False)

table_df["Average Purchase Price"] = table_df["Average Purchase Price"].map("${:.2f}".format)
table_df["Total Purchase Value"] = table_df["Total Purchase Value"].map("${:.2f}".format)

table_df = table_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

table_df.head()
```
